# ex 2
Create and label 4 separate charts with the following equations (choose a range for x that makes sense):
> y= âˆšx

> y= x^3

> y= 2^x

> y= 1/(x+1)

```python
import math
```

```python
x1 = range(10)
y1 = [math.sqrt(n) for n in x1]

x2 = range(10)
y2 = [(n)**3 for n in x1]

x3 = range(10)
y3 = [2**(n) for n in x1]

x4 = range(10)
y4 = [(1/(n+1)) for n in x1]
```

```python
plt.plot(x1,y1)
plt.show()

plt.plot(x2,y2)
plt.show()

plt.plot(x3,y3)
plt.show()

plt.plot(x4,y4)
plt.show()
```

```python
from random import randint, seed
seed(123)
```

```python
x1 = range(1,8)
y1 = [randint(0,5) + n for n in x1]

x2 = range(1,8)
y2 = [randint(0,10) + n for n in x2]

x3 = [1,2,3,4,5,6,7]
y3 = [randint(0,25) + m for m in x3]
```

```python
plt.plot(x1,y1, label = 'baby')
plt.plot(x2,y2, label = 'child')
plt.plot(x3,y3, label = 'adult')

plt.legend()
plt.show()
```
