```python
#1.Copy the users and roles DataFrames from the examples above.

# Create the users DataFrame.

users = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'name': ['bob', 'joe', 'sally', 'adam', 'jane', 'mike'],
    'role_id': [1, 2, 3, 3, np.nan, np.nan]
})
users

# Create the roles DataFrame

roles = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['admin', 'author', 'reviewer', 'commenter']
})
roles
```

```python
users
```

```python
#2. What is the result of using a right join on the DataFrames?
pd.merge(users, roles, how='right', left_on='role_id', right_on='id') #includes all rows from roles
```

```python
#3. What is the result of using an outer join on the DataFrames?
pd.merge(users, roles, how='outer', left_on='role_id', right_on='id') #includes all rows from both
```

```python
#4. What happens if you drop the foreign keys from the DataFrames and try to merge them?
pd.merge(users, roles, how='outer') #returned one df after another along similar keys
```

```python
#1. Use your get_db_url function to help you explore the data from the chipotle database.
from env import host, username, password
url_chipotle = get_db_url(username, host, password, 'chipotle')
orders = pd.DataFrame(pd.read_sql('select * from orders', url_chipotle))
orders
```

```python
def extract_price(string):
    cleaned_price = (string.replace('$', ''))
    return cleaned_price
```

```python
#2. What is the total price for each order?
orders['item_price'] = orders['item_price'].apply(extract_price) 
orders['item_price'] = orders['item_price'].astype(float)
```

```python
orders.groupby('order_id').item_price.sum()
```

```python
#3. What are the most popular 3 items?
orders.groupby('item_name').quantity.sum().nlargest(3)
```

```python
#4. Which item has produced the most revenue?
quantities=orders.groupby('item_name').quantity.sum()
orders.groupby('item_name').quantity.sum().nlargest(3)
```

```python
orders['rev_per_item']= orders.quantity * orders.item_price
orders.head()
```

```python
orders.groupby('item_name').rev_per_item.sum().nlargest(1)
```
